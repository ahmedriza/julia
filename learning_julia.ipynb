```julia
function mandelbrot(a)
    z = 0
    for i=1:50
        z = z^2 + a
    end
    return z
end
```

```julia
function test_manddelbrot()
    for y=1.0:-0.05:-1.0
        for x=-2.0:0.0315:0.5
            abs(mandelbrot(complex(x, y))) < 2 ? print("*") : print(" ")
        end
        println()
    end
end
```

```julia
test_manddelbrot()
```
